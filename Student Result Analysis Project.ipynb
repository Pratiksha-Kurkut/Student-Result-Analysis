```python
#drop unnamed column
df = df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

```python
#change weekly study hours column 
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("5 - 10","6 - 12")
df.head()
```

```python
#gender distribution
plt.figure(figsize= (2,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
#from the above chart we have analysed that:
#the number of females in the data is more than the number of males
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize= (4,4))
sns.heatmap(gb,annot = True)
plt.title("Relationship between Parent's Education And Student's Score")
plt.show()
```

```python
#from above chart we have concluded that,
#the education of parents have a good impact on their education.
```

```python
gb = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize= (3,3))
sns.heatmap(gb,annot = True)
plt.title("Relationship between Parent's Marital Status And Student's Score")
plt.show()
```

```python
#so the marital status of parents doesn't show major impact on the student's score.
```
